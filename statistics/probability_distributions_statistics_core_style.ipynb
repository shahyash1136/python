# 1. Basic Probability Distribution

## 📖 Concept

Easy explanation of **Basic Probability Distribution**.

### 🌍 Real Life Example
Dice rolling example

### 🎯 When To Use
- Understand probability behaviour
- Data science and ML applications


```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Topic: Basic Probability Distribution")
print("Example: Dice rolling example")
```


### 📊 Matplotlib Visualization


```python

x = np.arange(1,11)
y = np.random.randint(1,20,10)

plt.figure(figsize=(8,4))
plt.plot(x,y,marker='o')
plt.title("Sample Visualization")
plt.show()
```


### 📊 Seaborn Visualization


```python

sns.histplot(np.random.normal(size=1000), kde=True)
plt.show()
```


### 💡 Interpretation

- Graph ko observe karo
- Distribution ka shape identify karo
- Probability behaviour samjho
