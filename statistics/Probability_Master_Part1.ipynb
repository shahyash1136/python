```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()

df_batters = pd.DataFrame({
    "Player":["Virat","Gill","Rohit","KL","Pant","SKY","Rinku","Jaiswal","Iyer","Hardik"],
    "Runs":[741,890,417,520,446,605,474,625,530,350]
})

df_batters
```


# 1️⃣ Basic Probability

## What is Probability?

Probability kisi event ke hone ki chance ko measure karti hai.

Formula:

Probability = Favorable Outcomes / Total Outcomes

Range:
- 0 = Impossible
- 1 = Certain

Example:
- Dice par 6 aane ki probability = 1/6
- Coin par Head aane ki probability = 1/2


```python

success = (df_batters["Runs"] > 500).sum()
total = len(df_batters)

probability = success / total

print(f"Probability = {probability:.2f}")
```


### Interpretation

📊 Probability batati hai ki randomly ek batter choose karne par uske 500+ runs hone ki kitni chance hai.


```python

counts = [
    (df_batters["Runs"] > 500).sum(),
    (df_batters["Runs"] <= 500).sum()
]

labels = ["Runs > 500","Runs <= 500"]

plt.figure(figsize=(6,4))
plt.bar(labels, counts)
plt.title("Probability Example")
plt.show()
```


# 2️⃣ PMF (Probability Mass Function)

PMF discrete values ke liye use hota hai.

Question:

Randomly ek batter choose karo.

Uske runs 350, 417, 446... hone ki probability kya hai?


```python

pmf = df_batters["Runs"].value_counts(normalize=True).sort_index()

print(pmf)
```

```python

plt.figure(figsize=(10,4))
pmf.plot(kind="bar")
plt.title("PMF of Runs")
plt.ylabel("Probability")
plt.show()
```

```python

sns.barplot(x=pmf.index.astype(str), y=pmf.values)
plt.xticks(rotation=45)
plt.title("PMF using Seaborn")
plt.show()
```


### PMF Interpretation

PMF ki saari probabilities ka sum = 1 hota hai.

Discrete values:
- Number of wickets
- Number of sixes
- Number of matches



# 3️⃣ PDF (Probability Density Function)

PDF continuous data ke liye use hota hai.

Example:
- Height
- Weight
- Salary
- Bowling Average

PDF exact probability nahi deta.

Area under curve probability represent karti hai.


```python

heights = np.random.normal(170, 8, 1000)

height_df = pd.DataFrame({"Height": heights})

height_df.head()
```

```python

plt.figure(figsize=(8,4))
sns.histplot(height_df["Height"], kde=True)
plt.title("PDF Example - Heights")
plt.show()
```


### Interpretation

Bell shape dikhe to data approximately normal ho sakta hai.



# 4️⃣ CDF (Cumulative Distribution Function)

CDF batata hai:

X kisi value se chhota ya equal hone ki probability.

Example:

Height <= 170 ki probability?


```python

sorted_data = np.sort(heights)

cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)

plt.figure(figsize=(8,4))
plt.plot(sorted_data, cdf)
plt.title("CDF Example")
plt.xlabel("Height")
plt.ylabel("Cumulative Probability")
plt.show()
```


### Interpretation

CDF hamesha:

- 0 se start
- 1 par end

hoti hai.
